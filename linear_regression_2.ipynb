```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn import linear_model
```

```python
df = pd.read_csv(r"C:\Users\udayk\Downloads\homeprices5.csv")
```

```python
df
```

```python
%matplotlib inline
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red',marker='+')
```

```python
new_df = df.drop('price' , axis = 'columns')
new_df
```

```python
price = df.price 
price
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df, price )
```

```python
reg.predict([[3700]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
area_df = pd.read_csv(r"C:\Users\udayk\Downloads\areas.csv")
```

```python
p = reg.predict(area_df)
p
```

```python
area_df['prices']=p
area_df
```
